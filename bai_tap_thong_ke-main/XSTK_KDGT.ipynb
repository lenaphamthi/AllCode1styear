Giả thuyết null : Năng suất trung bình của một giống lúa ở các năm trước là 32,5

$$H_0 = 32,5 $$
Giả thuyết thay thế : Năng suất trung bình của một giống lúa ở các năm trước không phải là 32,5

$$H_a \ne 32,5 $$

```python
# Tạo hàm z_test:
def z_test(sample, hypoth_value, sig_level, variance, alternative):
    sample_mean = sample.mean()
    sample_st_dev = math.sqrt(variance)
    sample_size = len(sample)
    test_statistic = (sample_mean - hypoth_value) / \
                     (sample_st_dev/(math.sqrt(sample_size)))
    print(test_statistic)
    if alternative == "two-sided":
        pvalue = stats.norm.sf(np.abs(test_statistic)) * 2
    elif alternative == "larger":
        pvalue = stats.norm.sf(test_statistic)
    elif alternative == "smaller":
        pvalue = stats.norm.cdf(test_statistic)
    print(f'P Value = {pvalue}')
    if pvalue < sig_level:
        print("Có bằng chứng để bác bỏ giả thuyết H0")
    else:
        print("Chưa có bằng chứng để bác bỏ giả thuyết H0")
```

```python
# Tạo dữ liệu:
data = np.array([33.7, 35.4, 32.7, 36.3, 37.3, 32.4, 30.0,
        32.4, 31.7, 34.5, 42.0, 33.9, 38.1, 35.0, 33.8])
# Xác định các thông số:
mu0 = 32.5
sig = 0.01
var = 10
# Tiến hành kiểm định:
z_test(data, mu0, sig, var, "larger")
```

**Kết luận: Ta có bằng chứng để bác bỏ giả thuyết H0, tức là ý kiến khách hàng phản ánh đúng**

Ta xây dựng bài toán kiểm định:
$$H_0: σ^2 = 0.01$$
$$H_a: σ^2 > 0.01$$

```python
# Ta viết hàm kiểm định phương sai:
def var_test(var, df, va0, direction = "two-tailed", alpha = 0.05):
    Q = (df) * var / va0 
    if direction == "lower":
        q = stats.chi2.ppf(alpha, df)
        if Q <= q:
            print("Có bằng chứng để bác bỏ giả thuyết H0")
        else:
            print("Chưa có bằng chứng để bác bỏ giả thuyết H0")
    elif direction == "upper":
        q = stats.chi2.ppf(1 - alpha, df)
        if Q >= q:
            print("Có bằng chứng để bác bỏ giả thuyết H0")
        else:
            print("Chưa có bằng chứng để bác bỏ giả thuyết H0")
    else:
        q1 = stats.chi2.ppf(alpha / 2, df)
        q2 = stats.chi2.ppf(1 - (alpha / 2), df)
        if Q <= q1 or Q >= q2:
            print("Có bằng chứng để bác bỏ giả thuyết H0")
        else:
            print("Chưa có bằng chứng để bác bỏ giả thuyết H0")
```
