```python
import numpy as np
import matplotlib.pyplot as plt
```

image.png

- Đạo hàm bậc nhất : 3x^2 + 3
- Đạo hàm bậc hai : 6x 

image.png

```python
x = np.linspace(-6,5,num=1500) 
```

image.png

```python
y = x**3+ 3*x +7
y1 = 3*(x**2) + 3
y2 = 6*x 
```

image.png

image.png

```python
fig, ax = plt.subplots()
ax.plot(x,y,label="y(x)",color = "red")
ax.plot(x,y1,label="y'(x)",color = "orange")
ax.plot(x,y2,label="y''(x)",color="black")
plt.xlabel("x");
plt.ylabel("y");
plt.title("Đồ thị hàm số ")
plt.legend();
```
