6. Bạn hãy thử tìm hiểu các thao tác khác với ảnh như: (i) chuyển ảnh màu sang ảnh xám; (ii)
lưu ảnh vào máy; (iii) phòng to hay thu nhỏ ảnh; .v.v.

```python
#chuyển ảnh màu sang ảnh xám 
new_img = doc_img.convert("L")
new_img
```

```python
#lưu vào máy
new_img.save("new_totoro.jpg")
```

```python
#phóng to ảnh
new_width  = 600
new_height = 600
big_img = doc_img.resize((new_width,new_height))
big_img
```

```python
#thu nhỏ ảnh
new_height_ = 200
new_width_  = 200
small_img = doc_img.resize((new_height_,new_width_))
small_img
```

```python
#Chuyển ảnh màu sang Halftone 
new_img = doc_img.convert("1")
new_img 
```

```python
#làm mờ ảnh 
from PIL import ImageFilter
new_img = doc_img.filter(ImageFilter.GaussianBlur(radius=20))
new_img
```

```python
#xoay ảnh 
n_img = doc_img.rotate((180))
n_img
```

**Bài tập 2 - Thư viện thao tác với cấu trúc Tensor**

```python
#PyTorch 
import torch 
```

```python
#Tensor hạng 1 
ts_1 = torch.tensor([1,2,3,4,5])
print("Tensor hạng 1: ",ts_1)
print("hạng: ",ts_1.ndim)
```

```python
#Tensor hạng 2
ts_2 = torch.rand(2,2)
print("Tensor hạng 2: ")
print(ts_2)
print("hạng : ",ts_2.ndim)
```

```python
#Tensor hạng 3
ts_3 = torch.rand(2,3,5)
print("Tensor hạng 3: ")
print(ts_3)
print("hạng : ",ts_3.ndim)
```

```python
#Tensor hạng 4
ts_4 = torch.rand(1,3,4,6)
print("Tensor hạng 4: ")
print(ts_4)
print("hạng : ",ts_4.ndim)
```
