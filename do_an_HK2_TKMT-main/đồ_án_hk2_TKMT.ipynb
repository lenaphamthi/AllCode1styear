**Phân tích thống kê mô tả với dữ liệu định lượng** 
- Mthly_HH_Income : Thu nhập hàng tháng 

- Mthly_HH_Expense : Chi tiêu hàng tháng 

- Annual_HH_Income : Thu nhập hàng năm 

- Emi_or_Rent_Amt : Chi phí phát sinh 


```python
df_dl1 = pd.DataFrame(columns=["Mthly_HH_Income","Mthly_HH_Expense","Emi_or_Rent_Amt","Annual_HH_Income"],data=df)
```

Áp dụng các phép tính hướng tâm, phép tính phương sai – độ lệch chuẩn, các phép tính phân vị, cực đại – cực tiểu để tính các thông số cơ bản của các cột dữ liệu định lượng này.

```python
df_dl1.describe()
```

Ta bắt đầu tiến hành phân tích cho dữ liệu định lượng, phân tích cột Mthly_HH_Income & Mthly_HH_Expense để nhận biết mức độ chi phí và thu nhập của 20 dữ liệu đầu có cân bằng nhau không 

```python
#bảng thống kê loại thu nhập hàng tháng
type_hh_income = []
df1=df_dl1.copy()
for x in df1['Mthly_HH_Income']:
    if x <= 30000:
        type_hh_income.append('thu nhập Thấp')
    elif x <= 70000:
        type_hh_income.append('thu nhập Trung bình')
    else:
        type_hh_income.append('thu nhập Cao')
df1.insert(4, "Type_HH_Income", type_hh_income)
df1.head(20)
```

```python
#Bảng thống kê loại thu nhập hàng tháng tính ra phần trăm của cả tập dữ liệu  
dldl1 = df1.groupby('Type_HH_Income')['Type_HH_Income'].count()
dldl1 = pd.DataFrame(dldl1)
dldl1.columns = ['%']
dldl1 = dldl1/sum(dldl1['%'])*100
dldl1
```

```python
#Vẽ biểu đồ barplot cho bảng thống kê loại thu nhập hàng tháng
dldl1.reset_index(inplace=True)
sns.barplot(x='%',y='Type_HH_Income',data = dldl1 )
plt.title('Loại thu nhập', color = 'black', y = 1.1, fontsize = 20)
plt.show();
```

```python
#Bảng thống kê loại chi tiêu hàng tháng 
type_hh_expense = []
for x in df1['Mthly_HH_Expense']:
    if x <= 15000:
        type_hh_expense.append('Chi tiêu ít')
    elif x <= 30000:
        type_hh_expense.append('Chi tiêu trung bình')
    else:
        type_hh_expense.append('Chi tiêu cao')
df1.insert(5,"Type_HH_Expense", type_hh_expense)
df1.head(20)
```

```python
#Bảng thống kê loại chi tiêu hàng tháng tính ra phần trăm của tập dữ liệu  
dldl2 = df1.groupby('Type_HH_Expense')['Type_HH_Expense'].count()
dldl2 = pd.DataFrame(dldl2)
dldl2.columns = ['%']
dldl2 = dldl2/sum(dldl2['%'])*100
dldl2
```

```python
#Vẽ biểu đồ cho bảng thống kê loại chi tiêu hàng tháng
dldl2.reset_index(inplace=True)
sns.barplot(x='%',y='Type_HH_Expense',data = dldl2 )
plt.title('Loại chi tiêu', color = 'black', y = 1.1, fontsize = 20)
plt.show();
```

```python
#Vẽ biểu đồ boxplot cho tập dữ liệu thu nhập và chi tiêu hàng tháng để so sánh 
fig,(ax1,ax2)=  plt.subplots(nrows = 1,ncols = 2,figsize = (20,10))
ax1.boxplot(df["Mthly_HH_Income"]);
ax1.set(title = "Mthly_HH_Income (Thu nhập)");
ax2.boxplot(df["Mthly_HH_Expense"]);
ax2.set(title = "Mthly_HH_Expense (Chi tiêu)");
```

Phân tích cột Annual_HH_Income để xem hoàn cảnh gia đình sau 1 năm thu nhập, chi tiêu và chu trả cho những khoản phát sinh  

```python
#Bảng thống kê hoàn cảnh gia đình sau 1 năm 
type_situation = []
for x in df1['Annual_HH_Income']:
  if x <= 400000:
    type_situation.append('Thiếu thốn')
  elif x <=800000:
    type_situation.append('Khá Giả')
  else:
    type_situation.append('Đại gia')
df1.insert(6,"Type_Situation", type_situation)
df1.head(20)
```

```python
#Bảng thống kê hoàn cảnh gia đình sau 1 năm tính bằng phần trăm 
dldl3 = df1.groupby('Type_Situation')['Type_Situation'].count()
dldl3 = pd.DataFrame(dldl3)
dldl3.columns = ['%']
dldl3 = dldl3/sum(dldl3['%'])*100
dldl3
```

```python
#Vẽ biểu đồ cho bảng thống kê hoàn cảnh gia đình sau 1 năm 
dldl3.reset_index(inplace=True)
sns.barplot(x='%',y='Type_Situation',data = dldl3)
plt.title('Hoàn cảnh gia đình', color = 'black', y = 1.1, fontsize = 20)
plt.show();
```

```python
#Vẽ biểu đồ histogram cho tập dữ liệu 
fig,((ax1,ax2),(ax3,ax4)) =  plt.subplots(nrows = 2,ncols = 2,figsize = (40,10))
ax1.hist(df["Mthly_HH_Income"]);
ax1.set(title = "Mthly_HH_Income");
ax2.hist(df["Mthly_HH_Expense"]);
ax2.set(title = "Mthly_HH_Expense");
ax3.hist(df["Annual_HH_Income"]);
ax3.set(title = "Annual_HH_Income");
ax4.hist(df["Emi_or_Rent_Amt"]);
ax4.set(title = "Emi_or_Rent_Amt");
```
